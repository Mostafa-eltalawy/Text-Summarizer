```
models_to_run = []
```

## Baseline (NLTK Corpus)

```
from NLTK_summarizer import SummarizerNLTK
models_to_run.append({'model':SummarizerNLTK(), 'tag':'NLTK Corpus'})
```

## BERT

```
all_models = ['bert-base-uncased', 'bert-large-uncased', 'xlnet-base-cased', 'xlm-mlm-enfr-1024', 'distilbert-base-uncased', 'albert-base-v1', 'albert-large-v1']
all_clustering_algos = ['kmeans', 'gmm']

from BERT_summarizer import SummarizerBERT

for m in all_models:
    for a in all_clustering_algos:
        models_to_run.append({'model':SummarizerBERT(max_length=300, min_length=25, use_first=False, ratio=0.2, model=m,
                 clustering_algorithm=a), 'tag':'{} {}'.format(m,a)})
```

## T5 and Facebook BART

```
all_models = ['facebook/bart-large-cnn', 't5-11b', 't5-3b', 't5-base', 't5-large' ,'t5-small']

from T5_BART_summarizer import SummarizerT5BART

for m in all_models:
    models_to_run.append({'model':SummarizerT5BART(model=m, max_length=5, min_length=2), 'tag':'{}'.format(m)})
```

```
def runner(runNow, text):
    print("Starting")
    start = dt.now()
    try:
        result = runNow['model'].summary(text)
        done =True
        end = dt.now()
        timeTaken = (end-start).seconds
    except:
        result = "ERROR"
        done =False
        end = dt.now()
        timeTaken = (end-start).seconds
    print("DONE", result, done, timeTaken)
    return runNow, result, done, timeTaken
```

```
failed = ['t5-3b','t5-large'] # Failed multiple times
# failed = []
```

```
data = []
for i,m in enumerate(models_to_run):
    print('\n\n')
    pp(i+1)
    pp(m)
    if m['tag'] not in failed:
        runNow, result, done, timeTaken = runner(m, text)
    else:
        print("SKIPPING")
        runNow = m
        result = "ERROR"
        done = False
        timeTaken = 'SKIPPED'
    data.append({
        '#':i+1,
        'Model':runNow['tag'],
        'Time Taken(with Downloads)':timeTaken,
        'Completed':str(done),
        'Summary': result,
    })
```
